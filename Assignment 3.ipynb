```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
A = np.array([-3 , 0])
B = np.array([0 , -1.5])

X = ([A[0], B[0]])
Y = ([A[1], B[1]])

plt.plot(X, Y, 'r', label='(1  2)x=-3')

plt.plot(A[0] , A[1] , 'go' )
plt.plot(B[0] , B[1] , 'go' )

plt.text(A[0], A[1] , '(-3, 0)')
plt.text(B[0], B[1] , '(0, -3/2)')

plt.title("Plot of Straight Line")
plt.xlabel("X- Axis")
plt.ylabel("Y- Axis")
plt.grid(True)
plt.legend(loc='best')
plt.show()
```
