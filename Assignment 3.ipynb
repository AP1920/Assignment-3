```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,10,100)
y = (-x-3)/2
plt.plot(x, y, '-r', label='x+2y+3=0')
plt.title('Graph of x+2y+3=0')
plt.legend(loc='upper left')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.grid()
plt.show()
```
